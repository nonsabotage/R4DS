### First Step
mpgという車のデータを利用して, ggplot2を体験してみよう. 
とにもかくにもプロットをしてみると, displとhwyは負の関係がありそうなことがわかる. 

```r
print (mpg)
```

```r
ggplot(data=mpg) + 
    geom_point(mapping = aes(x = displ, y = hwy)) + 
    theme_classic()
```

```r
ggplot(data=mpg) + 
    geom_point (mapping = aes(x = displ, y =hwy, color= class)) + 
    theme_classic()
```

### facets
サブプロットを作成する際にとても有用. 

```r
ggplot(data=mpg) + 
    geom_point(mapping = aes(x=displ, y =hwy)) + 
    facet_grid(drv ~ cyl) + 
    theme_classic()
```

### Statistical Transformation

```r
# カウントまでを一緒に処理してくれる
ggplot(data=diamonds)+
    stat_count(mapping=aes(x=cut))+
    theme_classic()
```

```r
# データの値そのものをプロットしたい場合
demo <- tribble(
    ~a, ~b,
    "bar_1", 20, 
    "bar_2", 30, 
    "bar_3", 40
)
ggplot(demo) +
    geom_bar (
        mapping = aes(x = a, y = b), stat="identity"
    ) + theme_classic()
```

```r
# 比率をプロットしたい場合
p <- 
    ggplot(data=diamonds) + 
    geom_bar (
        mapping = aes(x = cut, y = ..prop.., group= 1)
    ) + 
    theme_classic()
p
```

```r
# プロットに関する値がない?
# 毎回計算しているのかも...
str(p)
```

```r
# summaryをプロットすることが可能
diamonds %>%
    ggplot() + 
    stat_summary (
        mapping = aes(x=cut, y = depth), 
        fun.ymin = min, 
        fun.ymax = max, 
        fun.y    = median
    )  + 
    geom_point(
        data = diamonds %>% group_by (cut) %>% summarise(depth = mean(depth)) %>% ungroup, 
        mapping = aes(x = cut, y = depth), 
        color = "red", 
        size  = 3
    )
```

### position adjusmtments

- identify
- fill
- dodge

```r
diamonds %>%
    ggplot() + 
    geom_bar (
        aes (x = cut, fill = clarity), 
        position = "fill"
    ) + 
    theme_classic()

diamonds %>%
    ggplot() + 
    geom_bar (
        aes (x = cut, fill = clarity), 
        position = "dodge"
    ) + 
    theme_classic()
```
