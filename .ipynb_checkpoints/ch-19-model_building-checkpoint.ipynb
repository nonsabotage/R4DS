## Why Low-Qualty Diamonds More Expensive

```r
b <- ggplot(diamonds)
b + geom_boxplot(aes(cut, price))
b + geom_boxplot(aes(color, price))
b + geom_boxplot(aes(clarity, price))
```

```r
ggplot(diamonds, aes(carat, price))+ 
    geom_hex(bins=50)
```

```r
## , 
diamonds2 <- diamonds %>%
    filter(carat <= 2.5) %>%
    mutate (lprice= log2(price), lcarat = log2(carat))

ggplot(diamonds2, aes(lcarat, lprice)) + 
    geom_hex(bins=50)
```

```r
## 
mod_diamonds2 <- lm (lprice ~ lcarat, data = diamonds2)
grid <- diamonds2 %>%
    data_grid(carat = seq_range(carat, 20)) %>%
    mutate (lcarat = log2 (carat)) %>%
    add_predictions(mod_diamonds2, "lprice") %>%
    mutate (price = 2 ** lprice)

ggplot(diamonds2, aes(carat, price)) + 
    geom_hex(bins=50) + 
    geom_line(data=grid, color="red", size=1)
```

```r
diamonds2 <- diamonds2 %>%
    add_residuals(mod_diamonds2, "lresid")
ggplot(diamonds2, aes(lcarat, lresid)) + 
    geom_hex(bins=50)
```

```r
## caratprice
## clarity. 
b <- ggplot(diamonds2)
b + geom_boxplot(aes(cut, lresid))
b + geom_boxplot(aes(color, lresid))
b + geom_boxplot(aes(clarity, lresid))
```

## A More Complicated Model

```r
mod_diamonds2 <- lm (
    lprice ~ lcarat + color + cut + clarity, data=diamonds2
)
```

```r
grid <- diamonds2 %>%
    data_grid(cut, .model = mod_diamonds2) %>%
    add_predictions(mod_diamonds2)
grid
```

```r
ggplot(grid, aes(cut, pred)) + 
    geom_point()
```

```r
diamonds2 <- diamonds2 %>%
    add_residuals(mod_diamonds2, "lresid2")
diamonds2 %>% print (n = 10)
```

```r
ggplot(diamonds2, aes(lcarat, lresid2)) + 
    geom_hex(bins=50)
```

## What Affects the Number  of Daily Flights?

```r
daily <- 
    flights %>%
    mutate (date = make_date(year, month, day)) %>%
    group_by (date) %>%
    summarise(n = n())
```

```r
ggplot(daily, aes(date, n)) + 
    geom_line()  +
    theme_classic()
```

#### Day of Week


```r
daily <- daily %>%
    mutate (wday = wday(date, label=TRUE))
ggplot(daily, aes(wday, n)) + 
    geom_boxplot()
```

```r
## 
## 
mod <- lm (n ~ wday, data=daily)
grid <- daily %>%
    data_grid(wday) %>%
    add_predictions(mod, "n")
ggplot(daily, aes(wday, n)) +
    geom_boxplot() + 
    geom_point (data=grid, color="red", size=4)
```

```r
## 
daily <- daily %>%
    add_residuals(mod)
daily %>%
    ggplot(aes(date, resid)) + 
    geom_ref_line(h = 0 ) + ## reference line
    geom_line()
```

```r
ggplot(daily, aes(date, resid, color=wday)) + 
    geom_ref_line(h = 0) + 
    geom_line() 
```

```r
## 
daily %>%
    filter (resid < -100)
```

```r
daily %>%
    ggplot(aes(date, resid)) + 
    geom_ref_line ( h= 0) + 
    geom_line(color="grey50") + 
    geom_smooth(se=FALSE, span=.2)
```

#### Seasonal Saturday Effect


```r
## , 
daily %>%
    filter (wday == "Sat") %>%
    ggplot (aes(date, n)) + 
        geom_point () + 
        geom_line () +
        scale_x_date (
            NULL, 
            date_breaks = "1 month", 
            date_labels = "%b"
        )
```

```r
## 3
term <- function (date) {
    cut (date, breaks = ymd(20130101, 20130605, 20130825, 20140101), 
        labels = c ("spring", "summer", "fall"))
}
daily <- daily %>%
    mutate (term = term (date))
daily %>% 
    filter (wday == "Sat") %>%
    ggplot(aes (date, n, color=term)) + 
    geom_point(alpha=1/3)  +
    geom_line ()  +
    scale_x_date (
        NULL, 
        date_breaks = "1 month", 
        date_labels = "%b"
    )



```
