### Duplicated Keys
keysが重複している場合にはすべての組み合わせが, 残る. 

```r
x <- tibble (
    key = c (1, 2, 2, 3), 
    val_x = paste0("x", 1:4)
)
y <- tibble (
    key   = c (1, 2, 2, 3), 
    val_y = paste0("y", 1:4)
)

left_join (x, y , by = "key")

```

## Set Operations

- intersect
- union
- setdiff


```r
df1 <- tibble (
    x = c (1, 2), 
    y = c (1, 1)
)
df2 <- tibble (
    x = c (1, 1), 
    y = c (1, 2)
)

intersect(df1, df2)
union(df1, df2)
setdiff(df1, df2)
```
